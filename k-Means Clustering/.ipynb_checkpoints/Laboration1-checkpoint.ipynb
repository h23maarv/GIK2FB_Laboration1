```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
file_path = "customerdata4.csv"
data = pd.read_csv(file_path)
```

```python
print(data.head())
print(data.describe())
```

```python
# Boxplot för Age
plt.subplot(1, 3, 1)
sns.boxplot(x="Gender", y="Age", data=data)
plt.title("Boxplot of Age by Gender")
plt.xlabel("Gender")
plt.ylabel("Age")
```

```python
# Boxplot för Annual Income
plt.subplot(1, 3, 2)
sns.boxplot(x="Gender", y="Annual Income (10kkr)", data=data)
plt.title("Boxplot of Annual Income by Gender")
plt.xlabel("Gender")
plt.ylabel("Annual Income (10kkr)")
```

```python
# Boxplot för Spending Score
plt.subplot(1, 3, 3)
sns.boxplot(x="Gender", y="Spending Score (1-100)", data=data)
plt.title("Boxplot of Spending Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Spending Score (1-100)")
```

```python
# Kolumner för analys (i detta fall Age och Spending Score)
point = data[["Age", "Spending Score (1-100)"]].values

# Separera x- och y-koordinater
xCoord = point[:, 0]
yCoord = point[:, 1]

# Visa data
print(xCoord)
print(yCoord)
```

```python
# Datapunkter
labels = range(1, len(point) + 1)

plt.figure(figsize=(8, 8))
plt.scatter(xCoord, yCoord, label='True Position')

plt.title("Data Points")
plt.xlabel("Age")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

```python
# Sätt antal kluster
k = 4  # Ändra beroende på uppgiften
kmeans = KMeans(n_clusters=k, random_state=42).fit(point)

# Hämta centroider och etiketter
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

print("Centroids:")
print(centroids)

# Visualisera klustringen
plt.figure(figsize=(8, 8))
plt.scatter(point[:, 0], point[:, 1], c=labels.astype(float), s=50, alpha=0.5, cmap="viridis")
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='x', label="Centroids")
plt.title("K-means Clustering")
plt.xlabel("Age")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.show()
```
